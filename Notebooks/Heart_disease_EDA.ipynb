# Data cleaning

## Things to do
1. Encoding
2. data conversion

```python
df.head()
```

```python
## encoding the categorical features
df['smoking']=df['smoking'].map({'Yes':1,'No':0})
```

```python
df.head()
```

```python
df['physical_activity']=df['physical_activity'].map({'Regular':2,'Rare':1,'No':0})
```

```python
df.head()
```

```python
df['alcohol']=df['alcohol'].map({'Yes':1,'No':0})
```

```python
df.head()
```

```python
df['heart_disease']=df['heart_disease'].map({'Yes':1,'No':0})
```

```python
df.head()
```

```python
df.info()
```

```python
df['smoking'].value_counts()
```

```python
df['alcohol'].value_counts()
```

```python
df['physical_activity'].value_counts()
```

```python
## lets save the cleaned data
df.to_csv("heart_disease_cleaned.csv", index=False)
```

### Lets start with EDA and Visualization

# EDA & Visualization


```python
plt.style.use('seaborn-v0_8')
df.hist(figsize=(20,15))
plt.show()
```

```python
## correlation 
df.corr()
```

```python
sns.heatmap(df.corr(),annot=True,cmap='coolwarm')
```

```python
## ploting piechart 
df['heart_disease'].value_counts().plot.pie(autopct="%1.1f%%")
mylabels=['Heart Disease','No Heart Disease']
plt.title("Heart Disease Distribution")
plt.legend(labels=mylabels)
plt.show()

```

```python
## analyzing alcohol consumption with heart disease
sns.countplot(x='alcohol',hue='heart_disease',data=df)
mylabels=['No Alcohol','Alcohol']
plt.legend(labels=mylabels)
plt.title("Alcohol Consumption vs Heart Disease")
plt.show()
```

This count plot shows a higher number of heart disease cases among alcohol consumers 
but it does not establish causation.

```python
##analyzing smoking consumption with heart disease
sns.countplot(x='smoking',hue='heart_disease',data=df)
mylabels=['No Smoking','Smoking']
plt.legend(labels=mylabels)
plt.title("Smoking Consumption vs Heart Disease")
plt.show()
```
